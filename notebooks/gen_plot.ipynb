```python
import matplotlib.pyplot as plt
```

```python
def plot_loss_curves(train_loss, val_loss):
    epochs = range(1, len(train_loss) + 1)
    
    plt.figure(figsize=(10, 6))
    plt.plot(epochs, train_loss, 'b-', label='Training Loss')
    plt.plot(epochs, val_loss, 'r-', label='Validation Loss')
    plt.title('Llama-3.2-3B Fine-Tuned Without CL (Math)', fontsize=18)
    plt.xlabel('Epoch', fontsize=14)
    plt.ylabel('Loss', fontsize=14)
    plt.legend(fontsize=14)
    plt.grid(False)
    plt.show()
```

```python
def plot_loss_curves_cl(train_loss, val_loss):
    epochs = range(1, len(val_loss) + 1)
    stage_length = len(val_loss) // 3
    
    plt.figure(figsize=(10, 6))
    plt.plot(epochs, train_loss, 'b-', label='Training Loss')
    plt.plot(epochs, val_loss, 'r-', label='Validation Loss')
    
    colors = ['lightgreen', 'yellow', 'salmon']
    labels = ['Easy', 'Medium', 'Hard']
    for i in range(3):
        start = i * stage_length
        end = (i + 1) * stage_length
        plt.axvspan(start + 0.5, end + 0.5, alpha=0.3, color=colors[i], label=labels[i])
    
    plt.title('Llama-3.2-3B Fine-Tuned With CL (Math)', fontsize=18)
    plt.xlabel('Epoch', fontsize=14)
    plt.ylabel('Loss', fontsize=14)
    plt.legend(fontsize=14)
    plt.grid(False)
    plt.show()
```
