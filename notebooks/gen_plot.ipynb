```python
import matplotlib.pyplot as plt
```

```python
def plot_loss_curves(train_loss, val_loss, subject):
    epochs = range(1, len(train_loss) + 1)
    
    plt.figure(figsize=(10, 6))
    plt.plot(epochs, train_loss, 'b-', label='Training Loss')
    plt.plot(epochs, val_loss, 'r-', label='Validation Loss')
    plt.title(f'Llama-3.2-3B Fine-Tuned Without CL ({subject})', fontsize=18)
    plt.xlabel('Epoch', fontsize=14)
    plt.ylabel('Loss', fontsize=16)
    plt.xlim(1, len(train_loss))
    plt.legend(fontsize=14)
    plt.grid(False)
    plt.show()
```

```python
train_loss = [7.1156, 3.3979, 2.2675, 1.8148, 1.5992, 1.4576, 1.3508, 1.2629, 1.2319, 1.1972]
val_loss = [3.8111793994903564, 2.475131034851074, 1.884594440460205, 1.5729366540908813, 1.448184847831726, 1.3845361471176147, 1.343237042427063, 1.317134976387024, 1.308098316192627, 1.303831934928894]
plot_loss_curves(train_loss, val_loss, 'Physics')
```

```python
def plot_loss_curves_cl(train_loss, val_loss, subject):
    epochs = range(1, len(val_loss) + 1)
    stage_length = len(val_loss) // 3
    
    plt.figure(figsize=(10, 6))
    plt.plot(epochs, train_loss, 'b-', label='Training Loss')
    plt.plot(epochs, val_loss, 'r-', label='Validation Loss')
    
    colors = ['lightgreen', 'yellow', 'salmon']
    labels = ['Easy', 'Medium', 'Hard']
    for i in range(3):
        start = i * stage_length
        end = (i + 1) * stage_length
        plt.axvspan(start + 0.5, end + 0.5, alpha=0.3, color=colors[i], label=labels[i])
    
    plt.title(f'Llama-3.2-3B Fine-Tuned With CL ({subject})', fontsize=18)
    plt.xlabel('Epoch', fontsize=14)
    plt.ylabel('Loss', fontsize=16)
    plt.xlim(1, len(val_loss))
    plt.legend(fontsize=14)
    plt.grid(False)
    plt.show()
```
